```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
print(stopwords.words('english'))
```

```python
#create function to remove stop words
def sw_removal(sentence):
    token_words=word_tokenize(sentence)
    token_words
    
    tokens_without_sw = " ".join([word for word in token_words if not word in stopwords.words()])
    
    return tokens_without_sw
```

```python
#apply sw_removal function on the nopunct_text column and create a new column for the text without SW
df["noSW_text"] = df["nopunct_text"].apply(lambda x: sw_removal(x))
```

We also chose to try stemming the sentences, to see whether this will ease the task of the algorithm when it will be classifying the text.

We conduct the stemming on the column with the lower case text without punctuation. This is because when conducting stemming things such as punctuation, capital letters, and stop words are not needed. We are trying to simplify sentences to a maximum, keeping only relevant information/words.

For the stemming to work, we first need to separate each word in the sentence (if not the function recognises the sentence as one word and does not change anything). To separate the sentence into words, we use tokenizer. The nltk tokenizer separates the sentence into words.

```python
from nltk.tokenize import sent_tokenize
from nltk.stem import PorterStemmer
```

```python
#create stemming function
porter=PorterStemmer()

def stem(sentence):
    token_words=word_tokenize(sentence)
    token_words
    stem_sentence=[]
    for word in token_words:
        stem_sentence.append(porter.stem(word))
        stem_sentence.append(" ")
    return "".join(stem_sentence)
```

```python
#apply stemming function on the lower_text column and create a new column for the stemmed text
#df["stemmed_lower_text"] = df["lower_text"].apply(lambda x: stem(x))
```

```python
#apply stemming function on the nopunct_text column and create a new column for the stemmed text
df["stemmed_nopunct_text"] = df["nopunct_text"].apply(lambda x: stem(x))
```

Another method we try is lemmatizing the sentences, also to see whether this will ease the task of the algorithm when it will be categorizing the text.

Before lemmatizing, we need to process to POS tagging. Otherwise each word is recognised as a noun and the lemmatization will not work effectively.

We start with the basic POS tagging.

We do the POS tagging (and then lemmatization) on the column with punctuated sentence, because working not punctuated sentence will lickely change the POS tagging into a wrong one. REFERENCE

```python
from nltk.corpus import wordnet as wn
from nltk.stem.wordnet import WordNetLemmatizer
from nltk import pos_tag
from collections import defaultdict
```

```python
tag_map = defaultdict(lambda : wn.NOUN)
tag_map['J'] = wn.ADJ    
tag_map['V'] = wn.VERB
tag_map['R'] = wn.ADV
```

```python
def lemma(sentence):
    token_words = word_tokenize(sentence)
    lemma_sentence=[]
    lemmatizer = WordNetLemmatizer()

    for token, tag in pos_tag(token_words):
        lemma_sentence.append(lemmatizer.lemmatize(token, tag_map[tag[0]]))
        lemma_sentence.append(" ")
    return "".join(lemma_sentence)
    #return tag_map[tag[0]]
```

```python
#apply POS tagging function on the text column and create a new column for the tagged text
df["lemmatized_text"] = df["lower_text"].apply(lambda x: lemma(x))
```
